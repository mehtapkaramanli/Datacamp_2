```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt; plt.rcdefaults()
import warnings
warnings.filterwarnings('ignore')
df = pd.read_csv('states_all.csv')
```

```python
math8=list()
states=list()
df.dropna().groupby('STATE').mean()
math8=df.dropna().groupby(by='STATE').mean()['AVG_MATH_8_SCORE']
states=df.dropna().groupby(by='STATE').mean()['AVG_MATH_8_SCORE'].index
```

```python
y_pos = np.arange(len(states))

plt.figure(figsize=(25,10))
plt.bar(y_pos, math8, align='center', alpha=0.9)
plt.xticks(y_pos, states,rotation='vertical')
plt.ylabel('AVG MATH SCORE')
plt.title('STATE-AVG MATH 8 SCORE')
 
plt.show()
```
